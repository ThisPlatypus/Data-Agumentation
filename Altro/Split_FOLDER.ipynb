```python
import os
import shutil
import random

def replicate_and_split_folders(source_dir, dest_dir1, dest_dir2, split_percentage):
    """
    Replicate the folder structure from source_dir to dest_dir1 and dest_dir2,
    splitting the files between the two directories based on the specified split_percentage.

    Args:
    - source_dir (str): Source directory to replicate.
    - dest_dir1 (str): Destination directory 1.
    - dest_dir2 (str): Destination directory 2.
    - split_percentage (float): Percentage of files to put in dest_dir1. Should be between 0 and 1.

    Returns:
    - None
    """
    # Create destination directories if they don't exist
    os.makedirs(dest_dir1, exist_ok=True)
    os.makedirs(dest_dir2, exist_ok=True)

    # Walk through the source directory and replicate the folder structure
    for root, dirs, files in os.walk(source_dir):
        # Get relative path from source directory
        relative_path = os.path.relpath(root, source_dir)

        # Create corresponding directories in destination directories
        dest_path1 = os.path.join(dest_dir1, relative_path)
        dest_path2 = os.path.join(dest_dir2, relative_path)
        os.makedirs(dest_path1, exist_ok=True)
        os.makedirs(dest_path2, exist_ok=True)

        # Distribute files between dest_dir1 and dest_dir2 based on split_percentage
        c=0
        for file in files:
            c=c+1
            source_file_path = os.path.join(root, file)
            if c <= round(len(files)*split_percentage):
                shutil.copy(source_file_path, dest_path1)
            else:
                shutil.copy(source_file_path, dest_path2)
                
'''        for d in dirs:
            if 'ttack' not in d.lower():
                dest_nt_path1 = os.path.join(dest_path1, 'NT')
                dest_nt_path2 = os.path.join(dest_path2, 'NT')
                os.makedirs(dest_nt_path1, exist_ok=True)
                os.makedirs(dest_nt_path2, exist_ok=True)
                shutil.move(os.path.join(root, d), dest_nt_path1)
                shutil.move(os.path.join(root, d), dest_nt_path2)'''
```
